```python
import numpy as np
import plotly.graph_objects as go
```

```python
def Lissajous(n,nx,ny,nz,φx=1,φy=1,φz=1,tf=2*np.pi):
    t = np.linspace(0, tf, 300)      #Nudo
    r = np.linspace(0, 2*np.pi, n+1) #Nodos
    x,y,z = np.cos(nx*t + φx), np.cos(ny*t + φy), np.cos(nz*t + φz)
    a,b,c = np.cos(nx*r + φx), np.cos(ny*r + φy), np.cos(nz*r + φz)

    #Hago las parejas de antípodas
    lineas = [[i, i + n//2] for i in range(n//2) ]

    #Nudo
    nudo = [go.Scatter3d(x=x, y=y, z=z, mode='lines', line=dict(width=3, color='brown'))]

    #Agrego los puentes al nudo                      
    for m in lineas:
        nudo.append(go.Scatter3d(x=[ a[m[0]], a[m[1]] ],
                                 y=[ b[m[0]], b[m[1]] ],
                                 z=[ c[m[0]], c[m[1]] ], line=dict(width=3)))

    fig = go.Figure(data=nudo)
    fig.show()
```

```python
Lissajous(10,3,5,1,0.1,0.7)
```

```python
Lissajous(100,3,5,1,0.1,0.7)
```

```python
Lissajous(10,2,3,5)
```
